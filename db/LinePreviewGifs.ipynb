```python
import os, sys, subprocess
from natsort import natsorted
```

```python
lines_folder = "../../ZBrain2/app/assets/images/0-Lines"
output_folder = "../app/assets/images/linegifs"
MH_folders = [d for d in os.listdir(lines_folder) if 'MH_' in d]
print(MH_folders)
```

```python
for MH_fol in MH_folders:
    pics = natsorted(os.listdir(lines_folder + '/' + MH_fol))[::2]
    pics = [lines_folder + '/' + MH_fol + '/' + p for p in pics]
    pics = ' '.join(pics)
    create_gif = 'convert -scale 900x400 -rotate 180 -delay 1 -loop 0 {0} {1}/{2}.gif'.format(pics, output_folder, MH_fol)
    print(create_gif) 
    subprocess.check_output(create_gif, shell=True)
```
